```python
a=np.linspace(0,10,11)
b=a**4
x=np.arange(0,10)
y=2*x
print(a)
print(b)
print(x)
y
```

```python
fig=plt.figure()
#large set of axes
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b);
axes.set_xlim(0,8)
axes.set_ylim(0,8000)
axes.set_xlabel('a')
axes.set_ylabel('b')
axes.set_title('power of 4')

#small axes

axes2=fig.add_axes([0.5,0.5,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(2,4)
axes2.set_ylim(0,300)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title('axes 2 plt')

plt.show()
```

```python
fig=plt.figure(dpi=100,figsize=(5,3))
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('new_figure.png',bbox_inches='tight')
```

```python
a=np.linspace(0,10,11)
b=a**4
x=np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.5,hspace=0.4)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(5,5),dpi=150)
axes[0][0].plot(x,y)
axes[0][0].set_title('0,0 title')
axes[0][1].plot(a,b)
axes[0][1].set_xlabel('x label')
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.suptitle('total title',fontsize=20)
plt.tight_layout()
fig.savefig('subplots.png',bbox_inches='tight')
```
